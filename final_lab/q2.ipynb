# Display 3 Shades of Each Color
This program creates and displays 3 shades (from black to full color) for each of the following colors:

- White
- Red
- Green
- Blue
- Cyan
- Magenta
- Yellow

Libraries used: OpenCV, NumPy, Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
color_names = ['White', 'Red', 'Green', 'Blue', 'Cyan', 'Magenta', 'Yellow']
color_rgbs = [
    (255, 255, 255),
    (255, 0, 0),
    (0, 255, 0),
    (0, 0, 255),
    (0, 255, 255),
    (255, 0, 255),
    (255, 255, 0)
]
shades = [0, 0.5, 1]
plt.figure(figsize = (12, 10))
total_row = len(color_names)
total_col = len(shades)

for row, (name, rgb) in enumerate(zip(color_names, color_rgbs)):
    offset = (row * total_col)
    for col, shade in enumerate(shades):
        img_no = offset + col + 1
        img = np.zeros((100, 100, 3), dtype = np.uint8)
        shaded_rgb = tuple([int(shade * c) for c in rgb])
        img[:] = shaded_rgb
        plt.subplot(total_row, total_col, img_no)
        plt.imshow(img)
        plt.title(name)
        plt.axis(False)


plt.tight_layout()
# plt.show()
plt.savefig('./outputs/q2.png')
```
