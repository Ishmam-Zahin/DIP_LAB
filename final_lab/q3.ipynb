# Display Bit-Plane Slicing of an image

perform bit-plane slicing on a gray scale image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread('/home/zahin/Desktop/DIP_LAB/final_lab/images/peppers.png', cv2.IMREAD_GRAYSCALE)
width = img_gray.shape[0]
height = img_gray.shape[1]
```

```python
img_slices = np.zeros((8, width, height), dtype = np.uint8)
```

```python
plt.figure(figsize = (12, 8))
total_col = 4
total_row = int(np.ceil((len(img_slices) + 1) / total_col))
```

```python
plt.subplot(total_row, total_col, 1)
plt.imshow(img_gray, cmap = 'gray')
plt.title('original')
plt.axis(False)

for index, slice in enumerate(img_slices):
    img_name = f'plane-{index}'
    img_pos = index + 2
    slice = (img_gray >> index) & 1
    slice = slice * 255
    plt.subplot(total_row, total_col, img_pos)
    plt.imshow(slice, cmap = 'gray')
    plt.title(img_name)
    plt.axis(False)

plt.tight_layout()
# plt.show()
plt.savefig('./outputs/q3.png')
```
